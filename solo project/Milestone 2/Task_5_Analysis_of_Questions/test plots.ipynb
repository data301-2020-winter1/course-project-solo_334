```python
import numpy as np
from numpy.random import randn

import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams

import seaborn as sns

import pandas_profiling

# Settings to for nice plots
%matplotlib inline

#rcParams['figure.figsize']=5,4
#sns.set_style['whitegrid']

%config InlineBackend.figure_format = 'svg' 
plt.style.use('fivethirtyeight')

# Reading datasets into pandas
crypto = pd.read_csv('/Users/dawsoncitygeneralstore/school/Data 301/Solo_project_6008/solo project/data/raw/Cryptocurrency uptodate files/crypto_currency_index.csv')

# Selecting the 'Rank' , 'Name' and the 'Market_Cap' columns
market_cap_raw = crypto[['Rank','Name','Market_Cap']]

# Counting the number of values
# ... YOUR CODE FOR TASK 2 ...
market_cap_raw.count().head()
```

```python
# Filtering out rows without a market capitalization
cap = market_cap_raw.query('Market_Cap >0')

# Counting the number of values again
cap.count()
```

```python
#Declaring these now for later use in the plots
TOP_CAP_TITLE = 'Top 10 market capitalization'
TOP_CAP_YLABEL = '% of total cap'

# Selecting the first 10 rows and setting the index to name
cap10 = cap[:10].set_index('Name')

# Calculating market_cap_perc
cap10 = cap10.assign(market_cap_perc = lambda x: (x.Market_Cap / cap.Market_Cap.sum())*100)

# Plotting the barplot with the title defined above 
ax = cap10.market_cap_perc.plot.bar(title=TOP_CAP_TITLE)

# Annotating the y axis with the label defined above
ax.set_ylabel(TOP_CAP_YLABEL);
```
